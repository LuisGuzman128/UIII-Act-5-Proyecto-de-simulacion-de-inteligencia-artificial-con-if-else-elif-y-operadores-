```python
# 18/mayo/2026

# Programador:
# Lenguaje de programación, experiencia,
# tipo de proyecto, nivel de dificultad,
# recomendación de IA y estado del proyecto.

# ======================================================
# 💻 CBTis 128
# 📚 Programación Orientada a Objetos 2026
# 🤖 Especialidad: Inteligencia Artificial
# 👨‍🏫 Docente: Eliseo Nava
# 👨‍🎓 Alumno: Axel Guzman NL = 34
# 💻 Proyecto: Simulación de Inteligencia Artificial
# ======================================================

print("=========== SISTEMA IA PROGRAMADOR ===========")

nombre = input("Ingresa nombre del programador: ")
lenguaje = input("Ingresa lenguaje de programación: ")
experiencia = int(input("Ingresa años de experiencia (solo el número): "))
proyecto = input("Ingresa tipo de proyecto: ")
dificultad = input("Ingresa nivel de dificultad (bajo, medio, alto): ")

print("\n=========== RESULTADOS IA ===========")

# --------------------------------------------------
# Inteligencia artificial usando if
# --------------------------------------------------

if experiencia >= 5:
    print("✅ Programador con experiencia avanzada")

# --------------------------------------------------
# Inteligencia artificial usando if else
# --------------------------------------------------

if dificultad == "alto":
    print("⚠️ Proyecto complejo")
else:
    print("✅ Proyecto manejable")

# --------------------------------------------------
# Inteligencia artificial usando if elif else
# --------------------------------------------------

if lenguaje == "python":
    print("🤖 Recomendación IA: Utilizar Machine Learning")

elif lenguaje == "java":
    print("☕ Recomendación IA: Crear aplicación empresarial")

elif lenguaje == "javascript":
    print("🌐 Recomendación IA: Desarrollo web interactivo")

else:
    print("💻 Lenguaje válido para programación general")

# --------------------------------------------------
# Operadores lógicos y relacionales
# --------------------------------------------------

if experiencia >= 3 and dificultad == "alto":
    print("🔥 IA detecta programador apto para proyectos complejos")

elif experiencia < 3 or dificultad == "medio":
    print("📚 IA recomienda seguir practicando programación")

# --------------------------------------------------
# Resultado final
# --------------------------------------------------

if proyecto == "videojuego":
    print("🎮 Estado del proyecto: Desarrollo creativo")

elif proyecto == "pagina web":
    print("🌐 Estado del proyecto: Desarrollo web")

elif proyecto == "app movil":
    print("📱 Estado del proyecto: Aplicación móvil")

else:
    print("💻 Estado del proyecto: Proyecto general")

print("\n✅ Simulación de inteligencia artificial finalizada")
```
